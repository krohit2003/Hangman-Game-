```python
freq_table1 = np.zeros(28)
freq_table2 = np.zeros((28, 28))
print(freq_table1)
print(freq_table2)
```

```python
word="apple"
unique_letters = np.zeros(28)  # reset per word
for ch in word:
    a = ord(ch) - ord('a')     # convert letter â†’ index
    unique_letters[a] = 1  
```

```python
print(unique_letters)
for i in range(28): 
    # print(i)
    if unique_letters[i] == 1:
        freq_table1[i] += 1

print(freq_table1)
```

```python


# Example full_dictionary (raw words without markers yet)
full_dictionary = [
    "apple",
    "banana",
    "cat",
    "quick",
    "nation",
    "happiness"
]


for word in full_dictionary:
            word='{'+word+'|'
            
            # table1
            unique_letters = np.zeros(28)
            for ch in word:
                a = ord(ch) - ord('a')
                unique_letters[a] = 1
            for i in range(28): 
                if unique_letters[i] == 1 : ##--'apple' in dictionary; count 'p' only one time
                    freq_table1[i] = freq_table1[i] + 1

            # table 2
            i = 0
            while i + 1 < len(word):
                a = ord(word[i]) -ord('a')
                b = ord(word[i+1]) -ord('a')
                freq_table2[a, b] = freq_table2[a, b] + 1
                i = i + 1
                

print(freq_table1)
print(freq_table2)    
print(freq_table2[0,1])  # Example to print frequency of 'p'  
```
